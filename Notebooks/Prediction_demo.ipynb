```python
import joblib

# Path to your saved model
model_path = "/Users/alice/Desktop/finalproject/Model/best_model_VIX_Up.joblib"

# Load the model
loaded_model = joblib.load(model_path)
```

```python
sample = {
    "percentage_laid_off": 0.05,
    "Layoff_Index": 2.3,
    "industry": "Tech",
    "location": "USA",
    "stage": "Series B",
    "country": "United States",
    "year": 2023,
    "month": 4,
    "day": 15
}
```

```python
import pandas as pd

sample_df = pd.DataFrame([sample])

prediction = loaded_model.predict(sample_df)
probability = loaded_model.predict_proba(sample_df)

print("Prediction:", prediction)
print("Probability:", probability)
#0 mean VIX Down 1 mean Vix up
#[0.9918   0.0081]
#0 VIX Down / No Increase: 99.18%
#1 VIX Up  0.81%
```
